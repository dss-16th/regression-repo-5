```python
# 결측치 데이터 불러오기
data = pd.read_csv('lifeexpectancy.csv')
data
```

```python
data.info()
```

```python
import missingno as msno
msno.matrix(data)
```

```python
# gdp 컬럼 삭제, 더미 변수화
data.drop(['GDP', 'Unnamed: 0'],axis=1)
```

```python
data = pd.get_dummies(data, columns=['Status'])
data.drop(['Unnamed: 0','Country'],axis=1)
data.info()
```

```python
data=data.drop(['Status_Developed','Status_Developing'],axis=1)
data.info()
```

```python
# GDP 자료 
data['GDP_section'] = pd.cut(data.GDP, bins = [ 1.040000,519.839630,2009.395612,6831.020770,189464.583600],labels=['low','low_mid','high_mid','high'])
data
```

```python
# 데이터 형변환
data=data.drop(['section_low','section_low_mid','section_high_mid','section_high'],axis=1)
data = data.astype('float')
data.info()
```

```python
#히스토그램
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
data.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#상관계수 히스토그램
corr = data.corr()
corr['Lifeexpectancy'].sort_values(ascending=False)

correlation_matrix = corr.round(1)
plt.figure(figsize=(15,10))
sns.heatmap(data = correlation_matrix, annot=True, cmap='bwr')
plt.show()
```

```python
# 1차 데이터 분류
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('Lifeexpectancy', axis=1)
y = data['Lifeexpectancy']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=13)
```

```python
#모델 학습
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
# 결과
import statsmodels.api as sm
lm = sm.OLS(y_train, X_train).fit()
lm.summary()
```

```python
#예측
predictions = lm.predict(X_test)
predictions
```

```python
#실제 test 값 
y_test
```

```python
# 참값과 예측값의 직관적인 비교

sns.scatterplot(y_test, predictions);
plt.plot([40,90],[40,90],'r',ls='dashed',lw=3);
```

```python
# 모델 평가 
from sklearn.metrics import mean_squared_error

pred_tr = reg.predict(X_train)
pred_test = reg.predict(X_test)

rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))
rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

print('RMSE of Train Data : ',rmse_tr)
print('RMSE of Test Data : ',rmse_test)
```

```python
from sklearn import metrics

print('MSE', metrics.mean_squared_error(y_test, predictions))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
#성능 확인
plt.scatter(y_test,pred_test)

plt.plot([0,90],[0,90], 'r')
plt.show()
```
